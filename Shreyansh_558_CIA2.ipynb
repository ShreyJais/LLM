```python
import numpy as np
#Sinusoidal Embedding with Cosine and Sine
def pe(position, _2i, d):
    if position % 2 == 0:
        return np.sin(position / (10000 ** (2 * _2i / d)))
    else: #odd position
        return np.cos(position / (10000 ** (2 * _2i / d)))
```

```python
#importing requried modules
import nltk
nltk.download('punkt')
from nltk import word_tokenize
```

```python
Text = input("Enter the text: ")
tokens = nltk.word_tokenize(Text.lower())
print(tokens)
d = 1
for p in range(len(tokens)):
  for i in range(d):
    print(f"PE for p={p}, i={i} :\t",pe(p,i,d))
```

```python
import matplotlib.pyplot as plt

# Visualization for d = 1, 128, and 512
for d in [1, 128, 512]:
  pe_matrix = np.zeros((len(tokens), d))  # Initializing new matrix for PE
  for p in range(len(tokens)):
      for i in range(d):
          pe_matrix[p, i] = pe(p, i, d) #insert values of PE
  plt.imshow(pe_matrix, aspect='auto')
  plt.title(f"Positional Encoding for d={d}")
  plt.xlabel("Dimension")
  plt.ylabel("Token Position")
  plt.colorbar()
  plt.show()
```
